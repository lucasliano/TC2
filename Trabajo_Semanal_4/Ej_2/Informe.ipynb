## Implementaci贸n Computarizada

#### Importamos las librerias a utilizar

```python
import numpy as np
import matplotlib.pyplot as plt
from splane import pzmap, grpDelay, bodePlot, convert2SOS
from scipy import signal

# Esta es una liberia tomada de la comunidad [https://stackoverflow.com/questions/35304245/multiply-scipy-lti-transfer-functions?newreg=b12c460c179042b09ad75c2fb4297bc9]
from ltisys import *

# M贸dulos para Jupyter (mejores graficos!)
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize'] = [12, 4]
plt.rcParams['figure.dpi'] = 150 # 200 e.g. is really fine, but slower
```

```python
# Coeficientes Transferencias
wp = 1
Q_p = np.sqrt(2) / 2

# Ejercicio 1a
k = 2
wz = k * wp

#Ejercicio 1b
p = 0.5
```

## Definimos la transferencia a partir de la expresi贸n hallada


```python
# Genero la funci贸n transferencia T1 en S
den_1 = [1, (wp / Q_p), wp**2]
num_1 = [1, 0, wz**2]
T1 = ltimul(num_1, den_1);


den_2 = [1, wp / Q_p, wp**2]
num_2 = [1, (wp / Q_p)*p, wp**2]
T2 = ltimul(num_2, den_2);
```

## Diagrama de polos y ceros


```python
pzmap(T1, 1);
pzmap(T2, 1);
```

## Diagrama de Bode

```python
fig, ax = bodePlot(T1.to_ss(), 2);
```

```python
fig, ax = bodePlot(T2.to_ss(), 2);
```
