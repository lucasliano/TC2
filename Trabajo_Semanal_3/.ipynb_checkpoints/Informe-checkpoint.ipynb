## Implementación Computarizada

#### Importamos las librerias a utilizar

```python
import numpy as np
import matplotlib.pyplot as plt
from splane import pzmap, grpDelay, bodePlot, convert2SOS
from scipy import signal

# Esta es una liberia tomada de la comunidad [https://stackoverflow.com/questions/35304245/multiply-scipy-lti-transfer-functions?newreg=b12c460c179042b09ad75c2fb4297bc9]
from ltisys import *

# Módulos para Jupyter (mejores graficos!)
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize'] = [12, 4]
plt.rcParams['figure.dpi'] = 150 # 200 e.g. is really fine, but slower
```

#### Inicializamos las variables utilizadas

Cargamos los valores correspondientes a la trasnferencia pasabanda normalizada.

### Transferencia $T_{1}(s)$
\begin{equation}
    T_{1}(s) = \dfrac{s\ \dfrac{1}{q_{1}}}{s^2 + s\ \dfrac{1}{q_{1}} + 1}
\end{equation}

### Transferencia $T_{2}(s) = T_{21}(s) \cdot T_{22}(s)$
\begin{equation}
    T_{21}(s) = k \dfrac{s\ \dfrac{\omega_{21}}{q_{21}}}{s^2 + s\ \dfrac{\omega_{21}}{q_{21}} + \omega_{21}^2} 
\end{equation}

\begin{equation}
    T_{22}(s) = k \dfrac{s\ \dfrac{\omega_{22}}{q_{22}}}{s^2 + s\ \dfrac{\omega_{22}}{q_{22}} + \omega_{22}^2} 
\end{equation}

**Nota:** $q_{21} \approx q_{22}$

```python
# Coeficientes Transferencias
q_1 = 7.89
w_1 = 1

k = 3.701
q_21 = 17.3
w_21 = 1.107

q_22 = 17.3
w_22 = 0.902
```

## Definimos la transferencia a partir de la expresión hallada


```python
# Genero la función transferencia T1 en S
num_t1 = [w_1 / q_1, 0]
den_t1 = [1, w_1 / q_1, w_1**2]
T1 = ltimul(num_t1, den_t1);

# Genero la función transferencia T21 en S
num_t21 = [k * w_21 / q_21, 0]
den_t21 = [1, w_21 / q_21, w_21**2]
T21 = ltimul(num_t21, den_t21);

# Genero la función transferencia T22 en S
num_t22 = [k * w_22 / q_22, 0]
den_t22 = [1, w_22 / q_22, w_22**2]
T22 = ltimul(num_t22, den_t22);

# Defino T2
T2 = T21 * T22
```

## Diagrama de polos y ceros

### Transferencia $T_{1}(s)$

```python
pzmap(T1, 1);
```

### Transferencia $T_{2}(s)$, $T_{21}(s)$ y $T_{22}(s)$

```python
pzmap(T21, 1);
pzmap(T22, 1);
pzmap(T2, 1);
```

### Transferencia $H(S) = H_{1}(S) \cdot H_{2}(S)$
<div class="alert alert-success">
    <strong>Nota:</strong> Estamos utilizando una nueva librería (ver en los imports) que permite multiplicar funciones de transferencia.
</div>


```python
T = T1 * T2
pzmap(T, 1);
```

## Diagrama de Bode

### Transferencia $T_{1}(s)$

```python
fig, ax = bodePlot(T1.to_ss(), 2);
ax[0].set_xlim(1e-1,1e1)
ax[0].set_ylim(-100, 10)
ax[1].set_xlim(1e-1,1e1)
```

### Transferencia $T_{21}(s)$

```python
fig, ax = bodePlot(T21.to_ss(), 2);
ax[0].set_xlim(1e-1,1e1)
ax[0].set_ylim(-100, 10)
ax[1].set_xlim(1e-1,1e1)
```

### Transferencia $T_{22}(s)$

```python
fig, ax = bodePlot(T22.to_ss(), 2);
ax[0].set_xlim(1e-1,1e1)
ax[0].set_ylim(-100, 10)
ax[1].set_xlim(1e-1,1e1)
```

### Transferencia $T_{2}(s)$

```python
fig, ax = bodePlot(T2.to_ss(), 2);
ax[0].set_xlim(1e-1,1e1)
ax[0].set_ylim(-100, 10)
ax[1].set_xlim(1e-1,1e1)
```

### Transferencia $T(s) = T_{1}(s) \cdot T_{2}(s)$

```python
fig, ax = bodePlot(T.to_ss(), 2);
ax[0].set_xlim(1e-1,1e1)
ax[0].set_ylim(-100, 10)
ax[1].set_xlim(1e-1,1e1)
```
