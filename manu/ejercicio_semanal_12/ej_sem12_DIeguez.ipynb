Simulacion de la funcion transferencia del punto B

```python
##Verificacion de la simulacion circuital

import numpy as np 
import matplotlib.pyplot as plt
from splane import pzmap, bodePlot
from scipy import signal
```

```python
tf = signal.TransferFunction([3,0,7,0],[1,0,7,0,10]) ##Y11
w=np.logspace(-1,2,1000,True,10.0,None,0) #Donde voy evaluar mi funcion de transferencia para el bode
#Armo el Bode
w, mag, fase = signal.bode(tf,w,None);

#Muestro mi diagrama de Bode
plt.figure()
plt.grid()
plt.semilogx(w, mag)    # Bode magnitud

##plt.ylim([-15 , 15])    #Limites del grafico de magnitud eje y (magnitud)
plt.figure()
plt.grid()
plt.semilogx(w, fase)  # Bode 
plt.show()

pzmap(tf)
```
